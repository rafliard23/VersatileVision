```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

# Load image jj-102

img_scene = cv.imread('JJ102.jpg')
img_show = cv.cvtColor(img_scene, cv.COLOR_BGR2RGB)
plt.imshow(img_show)
```

```python
# Process pencarian contour

img_gray = cv.cvtColor(img_scene, cv.COLOR_BGR2GRAY)
ret, threshold = cv.threshold(img_gray, 20, 255, 0)

plt.imshow(threshold)
```

```python
# mencari contour
contours, ret = cv.findContours(threshold, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```python
# Looping mencari kotak contour
pc_detections = []

img_final = img_show

for count in contours:
    
    # Mencari luas area
    area = cv.contourArea(count)
    if area >= 480:
        x, y, w, h = cv.boundingRect(count)

        pc_detections.append([x, y, w, h])
        cv.rectangle(img_final, (x, y), (x + w, y + h), (255, 0, 0), 1)

plt.imshow(img_show)
```
